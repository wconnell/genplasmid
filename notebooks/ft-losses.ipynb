```python
import matplotlib.pyplot as plt
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

# Load the TensorBoard events file
event_file = '../results/finetune/20241011_195152-8265fed3-5298-4fe3-9250-4cbeafd83af1/logs/events.out.tfevents.1728676329.default.11589.1'  # Replace with your file path
event_acc = EventAccumulator(event_file)
event_acc.Reload()

# Extract scalar metrics
train_loss = event_acc.Scalars('train/loss')
eval_loss = event_acc.Scalars('eval/loss')

# Prepare data for plotting
train_steps = [x.step for x in train_loss]
train_values = [x.value for x in train_loss]

eval_steps = [x.step for x in eval_loss]
eval_values = [x.value for x in eval_loss]

# Plot the training and evaluation losses
plt.figure(figsize=(7, 4.5))
plt.plot(train_steps, train_values, label='Training Loss', marker='o', markersize=0.5)
plt.plot(eval_steps, eval_values, label='Evaluation Loss', marker='s', markersize=3)
plt.title('Training and Evaluation Loss over Steps')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()

# Print final losses for reference
print(f"Final training loss: {train_values[-1]:.4f}")
print(f"Final evaluation loss: {eval_values[-1]:.4f}")
```
